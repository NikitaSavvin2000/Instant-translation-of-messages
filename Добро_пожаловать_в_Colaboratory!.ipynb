```
# Ploting packages
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from sklearn.metrics import mean_squared_error
# Date wrangling
from datetime import datetime, timedelta
import numpy as np
# Data wrangling
from sklearn.model_selection import GridSearchCV
import pandas as pd
import matplotlib.dates as mdates
# The deep learning class
from deep_model import DeepModelTS
from sklearn.preprocessing import MinMaxScaler
# Reading the configuration file
import yaml
from scipy.interpolate import interp1d
# Directory managment
import os
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import matplotlib.ticker as ticker

import os
import glob
import yaml

# Reading the hyper parameters for the pipeline
conf_files = ['conf.yml']
for conf_file in conf_files:
    with open(conf_file) as file:
        conf = yaml.load(file, Loader=yaml.FullLoader)

        if not os.path.exists('result'):
            os.makedirs('result')
        file_name = os.path.basename(conf_file).split('.')[0]  # получаем имя файла без расширения
        # Creating the folder with the corresponding name
        folder_name = f"result/{file_name}_result"
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)

        d = pd.read_csv('output.csv', sep=';')
        d['Datetime'] = [datetime.strptime(x, '%d.%m.%Y %H:%M') for x in d['Datetime']]

        d.sort_values('Datetime', inplace=True)

        d = d.set_index('Datetime')
        d.index = pd.to_datetime(d.index)

        d = d.groupby(pd.Grouper(freq="H")).sum()
        d = d.reset_index() # Add this line to reset the index and create separate columns for datetime and DAYTON_MW
        d_scaled = d.copy()
        scaler = MinMaxScaler(feature_range=(0, 1))

        scaled_data = scaler.fit_transform(d_scaled[['DAYTON_MW']])
        d_scaled['DAYTON_MW'] = scaled_data
        # Sorting the values
        #d_comparative = d.loc[d['Datetime'] >= '2021-03-27']
        #d_comparative = d_comparative.loc[d_comparative['Datetime'] <= '2022-12-31']
        print(d_scaled)
        deep_learner = DeepModelTS(
            data=d_scaled,
            Y_var='DAYTON_MW',
            lag=conf.get('lag'),
            LSTM_layer_depth=conf.get('LSTM_layer_depth'),
            epochs=conf.get('epochs'),
            train_test_split=conf.get('train_test_split'), # The share of data that will be used for validation
            batch_size=conf.get('batch_size'),
            activation=conf.get('activation'),
            optimizer=conf.get('optimizer')
        )
        train_test_split_1 = conf.get('train_test_split')
        # Fitting the model

        model = deep_learner.LSTModel()

        # Making the prediction on the validation set
        # Only applicable if train_test_split in the conf.yml > 0
        yhat = deep_learner.predict()
        yhat = np.array(yhat)
        yhat = yhat.reshape(-1, 1)
        yhat = scaler.inverse_transform(yhat)

        if len(yhat) > 0:
            y_true = d.tail(len(yhat))['DAYTON_MW'].values
            y_pred = yhat
            # Constructing the forecast dataframe
            fc = d.tail(len(yhat)).copy()
            fc.reset_index(inplace=True)
            fc['forecast'] = yhat

            # Ploting the forecasts
            plt.figure(figsize=(12, 8))
            for dtype in ['DAYTON_MW', 'forecast']:
                if dtype == 'DAYTON_MW':
                    label = 'History data consumption'
                    color = 'darkblue'
                else:
                    label = 'Forecast data consumption'
                    color = 'orange'
                plt.plot(
                    'Datetime',
                    dtype,
                    data=fc,
                    alpha=0.8,
                    label=label,
                    color=color
                )
            plt.title('The comparative graph')
            plt.legend()
            plt.grid(which='major', linestyle='--', linewidth='0.5', color='gray', alpha=0.5)

            ax = plt.gca()
            ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

            # Set the x-axis tick formatter to the DateFormatter with the format 'mm.yy'
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%m.%y'))
            ax.xaxis.set_tick_params(rotation=45, labelsize=10)
            plt.savefig(f"{folder_name}/The comparative graph_2.png")
        # Forecasting n steps ahead

        plt.figure(figsize=(12, 8))
        plt.plot(
            'Datetime',
            'DAYTON_MW',
            data=d_comparative,
            alpha=0.8,
            label='example',
            color='b'
        )
        plt.title('The comparative graph')
        plt.legend()
        plt.grid(which='major', linestyle='--', linewidth='0.5', color='gray', alpha=0.5)
        ax = plt.gca()
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%m.%y'))
        ax.xaxis.set_tick_params(rotation=45, labelsize=10)
        plt.savefig(f"{folder_name}/The comparative graph.png")

        mse_val = round(mean_squared_error(fc['DAYTON_MW'], fc['forecast']), 2)
        mape_val = round(np.mean((np.abs(fc['DAYTON_MW'] - fc['forecast']) / fc['DAYTON_MW']) * 100), 2)
        mae_val = round(mean_absolute_error(fc['DAYTON_MW'], fc['forecast']), 2)

        fc_scaled = fc.copy()
        scaled_data_1 = scaler.fit_transform(fc_scaled[['DAYTON_MW']])
        fc_scaled['DAYTON_MW'] = scaled_data_1
        scaled_data_2 = scaler.fit_transform(fc_scaled[['forecast']])
        fc_scaled['forecast'] = scaled_data_2

        mse_val_scaled = round(mean_squared_error(fc_scaled['DAYTON_MW'], fc_scaled['forecast']), 2)
        mape_val_scaled = round(np.mean((np.abs(fc_scaled['DAYTON_MW'] - fc['forecast']) / fc_scaled['DAYTON_MW']) * 100), 2)
        mae_val_scaled = round(mean_absolute_error(fc_scaled['DAYTON_MW'], fc_scaled['forecast']), 2)
        # Forecasting n steps ahead
        print(f"MSE: {mse_val:.2f}")
        print(f"MAPE: {mape_val:.2f}%")
        print(f"MAE: {mae_val:.2f}")

        print(f"MSE_scaled: {mse_val_scaled:.2f}")
        print(f"MAPE_scaled: {mape_val_scaled:.2f}%")
        print(f"MAE_scaled: {mae_val_scaled:.2f}")

        # Creating the model using full data and forecasting n steps ahead
        deep_learner = DeepModelTS(
            data=d_scaled,
            Y_var='DAYTON_MW',
            lag=conf.get('lag'),
            LSTM_layer_depth=conf.get('LSTM_layer_depth'),
            epochs=conf.get('epochs'),
            batch_size=conf.get('batch_size'),
            activation=conf.get('activation'),
            optimizer=conf.get('optimizer')
        )

        # Fitting the model
        deep_learner.LSTModel()

        # Forecasting n steps ahead
        n_ahead = 645
        yhat = deep_learner.predict_n_ahead(n_ahead)
        yhat = [y[0][0] for y in yhat]
        yhat = np.array(yhat)
        yhat = yhat.reshape((n_ahead, 1))
        yhat = scaler.inverse_transform(yhat).ravel()

        # Constructing the forecast dataframe
        fc = d.tail(76).copy()
        fc['type'] = 'original'

        last_date = max(fc['Datetime'])
        hat_frame = pd.DataFrame({
            'Datetime': [last_date + timedelta(days=x + 1) for x in range(n_ahead)],
            'DAYTON_MW': yhat,
            'type': 'forecast'
        })

        fc = pd.concat([fc, hat_frame])
        fc.reset_index(inplace=True, drop=True)

        # Ploting the forecasts
        plt.figure(figsize=(12, 8))
        for col_type in ['original', 'forecast']:
            plt.plot(
                'Datetime',
                'DAYTON_MW',
                data=fc[fc['type'] == col_type],
                label=col_type
            )
        plt.title('The consumption forecast until the end of 2024')
        plt.legend()
        plt.grid(which='major', linestyle='--', linewidth='0.5', color='gray', alpha=0.5)

        ax = plt.gca()
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

        # Set the x-axis tick formatter to the DateFormatter with the format 'mm.yy'
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%m.%y'))
        ax.xaxis.set_tick_params(rotation=45, labelsize=10)
        plt.savefig(f"{folder_name}/The consumption comparative forecast until the end of 2024.png")


        plt.figure(figsize=(12, 8))
        for col_type in ['forecast']:
            plt.plot(
                'Datetime',
                'DAYTON_MW',
                data=fc[fc['type'] == col_type],
                label='Frorecast',
                color='Orange'
            )
        plt.title('The consumption forecast until the end of 2024')
        plt.legend(
        )
        plt.grid(which='major', linestyle='--', linewidth='0.5', color='gray', alpha=0.5)

        ax = plt.gca()
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

        # Set the x-axis tick formatter to the DateFormatter with the format 'mm.yy'
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%m.%y'))
        ax.xaxis.set_tick_params(rotation=45, labelsize=10)
        plt.savefig(f"{folder_name}/The consumption forecast until the end of 2024.png")
        #plt.show()

        # Creating the folder if it doesn't exist

        # Saving the variables
        with open(f"{folder_name}/variables.txt", "w") as f:
            f.write(f"MSE: {mse_val:.2f}\n")
            f.write(f"MAPE: {mape_val:.2f}%\n")
            f.write(f"MAE: {mae_val:.2f}\n")
            f.write(f"-------------------------------------------------\n")
            f.write(f"MSE_scaled: {mse_val_scaled:.2f}\n")
            f.write(f"MAPE_scaled: {mape_val_scaled:.2f}%\n")
            f.write(f"MAE_scaled: {mae_val_scaled:.2f}\n")
            f.write(f"-------------------------------------------------\n")
            f.write(f"lag: {deep_learner.lag}\n")
            f.write(f"LSTM_layer_depth: {deep_learner.LSTM_layer_depth}\n")
            f.write(f"epochs: {deep_learner.epochs}\n")
            f.write(f"batch_size: {deep_learner.batch_size}\n")
            f.write(f"train_test_split: {train_test_split_1}\n")
            f.write(f"activation: '{deep_learner.activation}'\n")
            f.write(f"optimizer: '{deep_learner.optimizer}'\n")

```
